```python
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.utils import to_categorical
```

```python
#chargement de données
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()
```

```python
#pretraiQtement
#on transform chaque matrice (28x28) en vecteur de taille(784,) et normalisation
x_train = x_train.reshape(-1, 784) / 255.0
x_test = x_test.reshape(-1, 784) / 255.0

#pour transformer une étiquette en vecteur binaire on fait:
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)
```

```python
# Modèle
model = models.Sequential([
    layers.Dense(128, activation='relu', input_shape=(784,)),
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax')
])
```

```python
# Compilation
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# Entrainement
model.fit(x_train, y_train,
          epochs=5,
          batch_size=64,
          validation_split=0.2)
```

```python
#test
model.evaluate(x_test, y_test)
```

```python
from tensorflow.keras.datasets import cifar10
```

```python
(x_train, y_train), (x_test, y_test) = cifar10.load_data()
```

```python
# Prétraitement
x_train = x_train.reshape(-1, 3072) / 255.0
x_test = x_test.reshape(-1, 3072) / 255.0

y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)
```

```python
# Modèle
model = models.Sequential([
    layers.Dense(512, activation='relu', input_shape=(3072,)),
    layers.Dense(256, activation='relu'),
    layers.Dense(10, activation='softmax')
])
```

```python
#compliler
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
#le modele
model.fit(x_train, y_train,
          epochs=10,
          batch_size=64,
          validation_split=0.2)
```

```python
#test
model.evaluate(x_test, y_test)
```

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
#chargement
data = load_breast_cancer()
X = data.data
y = data.target
```

```python
# Normalisation
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
# Sparation 80% train et 20% test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# Modèle
model = models.Sequential([
    layers.Dense(64, activation='relu', input_shape=(30,)),
    layers.Dense(32, activation='relu'),
    layers.Dense(1, activation='sigmoid')
])
```

```python
#compilation
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(X_train, y_train,
          epochs=10,
          batch_size=32)
```

```python
model.evaluate(X_test, y_test)
```

```python
from tensorflow.keras.datasets import boston_housing
from sklearn.preprocessing import StandardScaler
import numpy as np
```

```python
# Chargement de données
(x_train, y_train), (x_test, y_test) = boston_housing.load_data()
```

```python
# Normalisation
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
# Modèle
model = models.Sequential([
    layers.Dense(64, activation='relu', input_shape=(13,)),
    layers.Dense(32, activation='relu'),
    layers.Dense(1)
])
```

```python
#compilation
model.compile(optimizer='adam',
              loss='mean_squared_error',
              metrics=[tf.keras.metrics.RootMeanSquaredError()])
```

```python
#training
model.fit(x_train, y_train,
          epochs=50,
          batch_size=32)
```

```python
#test
model.evaluate(x_test, y_test)
```
